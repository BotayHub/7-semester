```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy as sp
from scipy import stats
from statistics import mean
from functools import reduce
import math
from scipy.optimize import curve_fit
```

```python
file = pd.ExcelFile("data.xlsx")
```

### Кремний

```python
fig, ax = plt.subplots()
E = file.parse(sheet_name = 'Si', header=0, usecols='A').values.flatten() #мэВ
U_rel = file.parse(sheet_name = 'Si', header=0, usecols='B').values.flatten()

plt.scatter(E, U_rel, color = 'black', s = 2)


plt.grid(True)
ax.minorticks_on()
ax.grid(which='major', linewidth = 0.75)
ax.grid(which='minor', linewidth = 0.5)
ax.set_ylabel('U/N')
ax.set_xlabel('E, мЭВ')
plt.show()
fig.savefig('Si.png', dpi = 1000)
```

```python
fig, ax = plt.subplots()
E = file.parse(sheet_name = 'Si_1', header=0, usecols='A').values.flatten() #мэВ
U_rel_sqrt = file.parse(sheet_name = 'Si_1', header=0, usecols='B').values.flatten()


plt.scatter(E, U_rel_sqrt, color = 'black', s = 2)

z = np.polyfit(E[11:21], U_rel_sqrt[11:21], 1)
f = np.poly1d(z)
f_new = np.linspace(E[11], E[21], 100)
l_new = f(f_new)

print(f.c) # вывели коэффициенты a и b соответственно (y = ax+b)

plt.plot(f_new, l_new, linewidth = 1, color = 'black', label = 'y = 2.1x -2.4')

plt.grid(True)
ax.minorticks_on()
ax.grid(which='major', linewidth = 0.75)
ax.grid(which='minor', linewidth = 0.5)
ax.set_ylabel('(U/N)^0.5')
ax.set_xlabel('E, мЭВ')
plt.legend()
plt.show()
fig.savefig('Si_add.png', dpi = 1000)
```

```python
print('Ширина запрещенной зоны:', (abs(f.c[1])/f.c[0])-50)
```
